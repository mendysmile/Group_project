# 資料清理

```python
#把全形字或標點符號轉成半形, 轉成半形之後的標點符號才可被isalpha()清掉
def conv_wide(text):
    code = ord(str(text))
    if x == u'\u3000': #space
        code = 32
    elif u'\uff01' <= text <= u'\uff5e':
        code -= 65248
    return chr(code)
```

```python
#articles去除所有的標點符號
from datetime import datetime
for k in range(len(articles)):
    temp = []
    print(k)
    try:
        for wd in str(articles[k]):
            x = wd.upper()
            y = conv_wide(x)
            if x.isalpha():
                temp.append(y)
            new_article = ''.join(temp)   
        articles[k] = new_article
    except Exception as e:
        print (e.__doc__) 
```

```python
import jieba
jieba.load_userdict(r'./userdict.txt')
for j in range(len(articles)):
    try:
        cut[j] = jieba.lcut(str(articles[j]), cut_all=False) #jieba.lcut 直接返回 list, 而不是 generator
    except Exception as e:
        print (e.__doc__)
    cut_list = list()
    for wd in cut[j]:
        if wd != ' ':
            cut_list.append(wd)
        #丟入tfidf的語料必須為string, 因此先用 ' '連接變成一篇文章再丟入
        cut[j] = " ".join(cut_list).lower()
```

# 使用TF-IDF挑出每篇文章的特徵字

```python
!pip install sklearn
from sklearn.feature_extraction.text  import  TfidfVectorizer
#將詞頻矩陣X統計成TF-IDF值
cv = TfidfVectorizer(min_df=.01, max_df=.55)
tf = cv.fit_transform(df_idf)
features = cv.get_feature_names()
```

```python
tf
```

```python
print(features)
```

```python
tfs_new = pd.DataFrame(tf.toarray(), columns = features)
len(tfs_new)
```

```python
tfs_new
```

```python
for k in range (0, len(tfs_new)):
    print(k)
    #各document,依照tf-idf的值,把其特徵值做排序
    x = tfs_new.iloc[k].sort_values(ascending=False)
    #只把特徵值的名字取出
    y = list(x.index)
    #取出排名前30個的特徵值取出
    new_set = set()
    for j in range(0, 30):
        new_set.add(y[j])         
    df_idf[k] = ' '.join(new_set)
```
